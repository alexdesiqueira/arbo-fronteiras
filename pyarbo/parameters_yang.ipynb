As funções abaixo foram criadas para gerar os dicionários salvos no arquivo parameters.py. Os valores dos parâmetros, em alguns casos, foram fixados abaixo ou acima de uma temperatura específica seguindo o intervalo utilizado nos experimentos do yang. 

```python
def theta_m(): 
    df = pd.read_csv('../data/parameters_model.csv')

    df['T'] = df['T'].round(1)

    df.set_index('T', inplace = True)

    theta_m_values = df[['extrinsic_encubation']].copy()
    
    del df 
    
    theta_m_values.loc[theta_m_values.index <= 14.0, 'extrinsic_encubation'] = 0
    
    return theta_m_values.to_dict()['extrinsic_encubation']
```

```python
def gamma_m(): 
    df = pd.read_csv('../data/parameters_model.csv')

    df['T'] = df['T'].round(1)

    df.set_index('T', inplace = True)

    gamma_m_values = df[['aquatic_transition']].copy()
    
    del df 
    
    gamma_m_values.loc[gamma_m_values.index <= 14.7, 'aquatic_transition'] = 0.014642
    gamma_m_values.loc[gamma_m_values.index >= 36.5, 'aquatic_transition'] = 0.148137
    
    return gamma_m_values.to_dict()['aquatic_transition']
```

```python
def mu_a():
    df = pd.read_csv('../data/parameters_model.csv')

    df['T'] = df['T'].round(1)

    df.set_index('T', inplace = True)

    mu_a_values = df[['aquatic_mortality']].copy()

    mu_a_values.loc[mu_a_values.index <= 10.0, 'adult_mortality'] = 0.209681
    mu_a_values.loc[mu_a_values.index >= 33.4, 'adult_mortality'] = 0.074976
    
    return mu_a_values.to_dict()['aquatic_mortality']
```
